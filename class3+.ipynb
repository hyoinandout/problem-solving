```python
#2667 단지번호붙이기
from collections import deque
n = int(input())
arr = [[0 for _ in range(n)] for _ in range(n)]
visited = [[0 for _ in range(n)] for _ in range(n)]
for i in range(n):
    arr[i] = input()

result = []
def bfs(starty, startx, visited,result):
    dx = [-1,1,0,0]
    dy = [0,0,-1,1]
    queuex = deque()
    queuey = deque()
    queuex.append(startx)
    queuey.append(starty)
    visited[starty][startx]=1
    cnt=1
    while queuex and queuey:
        x = queuex.popleft()
        y = queuey.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if(nx<n and nx>=0 and ny<n and ny>=0):
                if(visited[ny][nx]==0 and arr[ny][nx]=='1'):
                    queuex.append(nx)
                    queuey.append(ny)
                    visited[ny][nx]=1
                    cnt+=1
    result.append(cnt)

for i in range(n):
    for j in range(n):
        if(arr[i][j]=='1' and visited[i][j] == 0):
            bfs(i,j,visited,result)

result.sort()
print(len(result))
for i in range(len(result)):
    print(result[i])
```

```python
#6064 카잉달력
from math import gcd
def lcm(x,y):
    return x*y//gcd(x,y)

def cr(m,n,x,y):
    if m == x:
        x = 0
    i = x
    while i <= lcm(m,n):
        if(i%n==y):
            return i
        i+=m
    if n == y:
        y = 0
    if(x==y):
        return lcm(m,n)
    j = y
    while j <= lcm(m,n):
        if(j%m==x):
            return j
        j+=n
    return -1
                
t = int(input())
for _ in range(t):
    m,n,x,y = map(int,input().split())
    print(cr(m,n,x,y))
```

```python
#7662 이중순위 큐
from queue import PriorityQueue
t = int(input())
for _ in range(t):
    max_queue = PriorityQueue()
    min_queue = PriorityQueue()
    s = {}
    cnt = 0
    k = int(input())
    for _ in range(k):
        c,n = input().split()
        if(c=='I'):
            max_queue.put(-(int(n)))
            min_queue.put(int(n))
        elif(c=='D'):
            if(cnt == max_queue.qsize() + min_queue.qsize()):
                continue
            cnt+=1
            if(int(n)==1):
                x = -max_queue.get()
                if x in s:
                    s[x]+=1
                else:
                    s[x]=1
            elif(int(n)==-1):
                y = min_queue.get()
                if y in s:
                    s[y]+=1
                else:
                    s[y]=1
    if(cnt == max_queue.qsize() + min_queue.qsize()):
        print("EMPTY")
    else:
        x = -(max_queue.get()) 
        y = min_queue.get()
        while True:
            if (x not in s and y not in s) or (s[x]== 0 and s[y] == 0):
                print(x,y)
                break
            else:
                if s[x]!=1:
                    x = -(max_queue.get())
                    s[x]-=1
                else:
                    y = min_queue.get()
                    s[y]-=1
```

```python
from collections import deque
def bfs(a,b):
    deq = deque()
    deq.append([a,''])
    cnt = 0
    k = [False]* 10000
    while deq:
        z = deq.popleft()
        x,y = z[0],z[1]
        k[x] = True
        if x == b:
            return y
        i = x*2
        if(i>=10000):
            i = i%10000
        if(k[i]==False):
            k[i]=True
            deq.append([i,y+'D'])
        j = (x%1000)*10+x//1000
        if(k[j]==False):
            k[j]=True
            deq.append([j,y+'L'])
        l = (x%10)*1000+(x-(x%10))//10
        if(k[l]==False):
            k[l]=True
            deq.append([l,y+'R'])
        if x == 0:
            x = 9999
        else:
            x -= 1
        if(k[x]==False):
            k[x]=True
            deq.append([x,y+'S'])
t = int(input())
for _ in range(t):
    a,b = map(int,input().split())
    print(bfs(a,b))
```
